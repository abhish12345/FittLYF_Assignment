******2. Autoencoder (Optional Deep Learning Model)******

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from sklearn.preprocessing import StandardScaler
```

```python
# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_train)
```

```python
# Autoencoder architecture
input_dim = X_scaled.shape[1]
encoding_dim = 14  # Half the input size for compression
```

```python
input_layer = Input(shape=(input_dim,))
encoder = Dense(encoding_dim, activation="relu")(input_layer)
encoder = Dense(encoding_dim // 2, activation="relu")(encoder)
decoder = Dense(encoding_dim, activation="relu")(encoder)
decoder = Dense(input_dim, activation="sigmoid")(decoder)
```

```python
autoencoder = Model(inputs=input_layer, outputs=decoder)
autoencoder.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the Autoencoder
history = autoencoder.fit(X_scaled, X_scaled, epochs=10, batch_size=32, shuffle=True, validation_split=0.2, verbose=1)
```

```python
import numpy as np
# Predict the reconstruction loss
reconstructed_data = autoencoder.predict(X_scaled)
reconstruction_error = np.mean(np.power(X_scaled - reconstructed_data, 2), axis=1)
```

```python
# Threshold for anomalies
threshold = np.percentile(reconstruction_error, 99.7)  # Top 0.3% is treated as anomaly
y_pred_ae = [1 if e > threshold else 0 for e in reconstruction_error]
```

```python
# Evaluate Autoencoder
print(f"Precision: {precision_score(y_train, y_pred_ae):.4f}")
print(f"Recall: {recall_score(y_train, y_pred_ae):.4f}")
print(f"F1-Score: {f1_score(y_train, y_pred_ae):.4f}")
print(f"ROC-AUC: {roc_auc_score(y_train, y_pred_ae):.4f}")
```

3. Model Evaluation and Discussion
Here’s how the models can be evaluated in terms of precision, recall, F1-score, and ROC-AUC:

Precision: Measures how many of the predicted fraudulent transactions are actually fraud. A high precision means fewer false positives (i.e., fewer non-fraud transactions classified as fraud).
Recall: Measures how many of the actual fraudulent transactions were correctly identified. A high recall means fewer false negatives (i.e., fewer fraudulent transactions were missed).
F1-Score: The harmonic mean of precision and recall, giving a balance between the two.
ROC-AUC: Measures the model's ability to distinguish between classes. A high ROC-AUC indicates better overall classification performance.

4. Trade-offs in Detecting Frauds
False Positives (FP): A false positive occurs when a legitimate transaction is identified as fraudulent. This can lead to a bad user experience, as legitimate users may have their transactions flagged or blocked. In fraud detection, minimizing false positives is important but not as critical as minimizing false negatives.

False Negatives (FN): A false negative occurs when a fraudulent transaction is missed. This is much more severe, as missing a fraudulent transaction can lead to significant financial losses for both the user and the financial institution.

5. Model Interpretation
Isolation Forest: This model works well with outlier detection and can handle large amounts of data efficiently. It tends to have a higher recall, meaning it identifies more frauds but may also result in higher false positives.

Local Outlier Factor (LOF): LOF is sensitive to local anomalies, meaning it can capture fraud that is more subtle in nature, but it may not perform as well in detecting global anomalies.

Autoencoder (Deep Learning): This method works well with complex data and can automatically learn the data structure. However, it requires more computational power and can be harder to interpret compared to other traditional methods like Isolation Forest or LOF.

```python
def classify_transactions(model, new_data):
    """
    Classifies new credit card transactions as fraudulent or normal.
    
    Parameters:
    - model: trained anomaly detection model (e.g., IsolationForest)
    - new_data: new dataset of transactions (pandas DataFrame)

    Returns:
    - classified_data: DataFrame with predicted fraud labels (0: normal, 1: fraud)
    """
    # Preprocessing (assumes the 'Amount' and 'Time' features need scaling)
    new_data['Amount_scaled'] = scaler.transform(new_data['Amount'].values.reshape(-1,1))
    new_data['Time_scaled'] = scaler.transform(new_data['Time'].values.reshape(-1,1))

    # Drop original Amount and Time columns
    new_data_cleaned = new_data.drop(columns=['Amount', 'Time'])

    # Predict fraud using the trained model
    predictions = model.predict(new_data_cleaned)

    # Convert anomaly labels (-1 for fraud, 1 for normal) to 1 for fraud and 0 for normal
    new_data['Fraud'] = [1 if p == -1 else 0 for p in predictions]

    return new_data[new_data['Fraud'] == 1]  # Return only fraudulent transactions
```

```python
def classify_transactions(model, new_data):
    """
    Classifies new credit card transactions as fraudulent or normal.
    """
    # Preprocessing: scale Amount and Time
    new_data['Amount_scaled'] = scaler.transform(new_data['Amount'].values.reshape(-1,1))
    new_data['Time_scaled'] = scaler.transform(new_data['Time'].values.reshape(-1,1))

    # Drop original Amount and Time columns
    new_data_cleaned = new_data.drop(columns=['Amount', 'Time'])

    # Predict fraud using the trained model
    predictions = model.predict(new_data_cleaned)

    # Convert anomaly labels (-1 for fraud, 1 for normal) to 1 for fraud and 0 for normal
    new_data['Fraud'] = [1 if p == -1 else 0 for p in predictions]

    return new_data[new_data['Fraud'] == 1]  # Return only fraudulent transactions
```

```python
def visualize_anomalies(data, fraud_data):
    """
    Visualizes fraudulent transactions using PCA for dimensionality reduction.
    """
    from sklearn.decomposition import PCA

    # Apply PCA to reduce to 2 components
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(data.drop(columns=['Amount', 'Time', 'Fraud']))

    # Create DataFrame for PCA results
    pca_df = pd.DataFrame(data=X_pca, columns=['PC1', 'PC2'])
    pca_df['Fraud'] = data['Fraud']

    # Plot the anomalies
    plt.figure(figsize=(10,6))
    sns.scatterplot(x='PC1', y='PC2', hue='Fraud', data=pca_df, palette={0: 'green', 1: 'red'}, alpha=0.7)
    plt.title('Anomalies Detected (PCA)')
    st.pyplot(plt)

if __name__ == "__main__":
    main()
```
